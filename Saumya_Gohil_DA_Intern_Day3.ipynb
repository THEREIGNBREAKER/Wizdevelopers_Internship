```python
# Histogram of Age
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Bar plot of Gender distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Pie chart of Class distribution
class_counts = df['class'].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Class Distribution')
plt.show()

```

```python
# Box plot for performance metrics by Class
plt.figure(figsize=(15, 10))
sns.boxplot(x='class', y='gripForce', data=df)
plt.title('Grip Force by Class')
plt.xlabel('Class')
plt.ylabel('Grip Force (kg)')
plt.show()

```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='class', y='sit and bend forward_cm', data=df)
plt.title('Sit and Bend Forward by Class')
plt.xlabel('Class')
plt.ylabel('Sit and Bend Forward (cm)')
plt.show()

```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='class', y='sit-ups counts', data=df)
plt.title('Sit-ups Counts by Class')
plt.xlabel('Class')
plt.ylabel('Sit-ups Counts')
plt.show()

```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='class', y='broad jump_cm', data=df)
plt.title('Broad Jump by Class')
plt.xlabel('Class')
plt.ylabel('Broad Jump (cm)')
plt.show()
```

```python
sns.scatterplot(x='height_cm',y='weight_kg',data=df,hue='class',palette='coolwarm')
plt.xlim(135,195)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
# Central Tendency
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]

# Measure of Variance/Standard Deviation
std_values = df.std()
variance_values = df.var()

print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\nStandard Deviation:\n", std_values)
print("\nVariance:\n", variance_values)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['weight_kg'], kde=True, color='blue', binwidth=1)
plt.axvline(x=67, color='red', linestyle='--', label='Hypothesized Mean (67 kg)')
plt.title('Distribution of Weight with Hypothesized Mean')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
anova_result = stats.f_oneway(df['height_cm'], df['weight_kg'], df['body fat_%'])
print("ANOVA result:", anova_result)
```

```python
sns.displot(df, x="weight_kg", hue="class", kind="kde", multiple="stack")
```

```python
contingency_table = pd.crosstab(df['gender'], df['class'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-square test:")
print("Chi2:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python
t_stat, p_val = stats.ttest_1samp(df['height_cm'], 170)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```
